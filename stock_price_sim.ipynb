```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
S0 = 100        # initial stock price
mu = 0.05       # expected return
sigma = 0.2     # standard deviation of returns
T = 1           # time in years
dt = 0.001      # time step
N = int(T / dt) # number of time steps
t = np.linspace(0, T, N) # time vector

# Simulating the stock price path
W = np.random.standard_normal(size=N)
W = np.cumsum(W)*np.sqrt(dt) # cumulative sum to simulate Brownian motion
X = (mu - 0.5 * sigma**2) * t + sigma * W
S = S0 * np.exp(X) # geometric Brownian motion

# Plotting
plt.figure(figsize=(14, 8))
plt.plot(t, S)
plt.title('Simulated Geometric Brownian Motion')
plt.xlabel('Time (years)')
plt.ylabel('Stock Price')
plt.show()
```

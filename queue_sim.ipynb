```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Parameters
lambda_arrival = 1        # average number of arrivals per minute
lambda_service = 1.5      # average number of services completed per minute
num_minutes = 120         # simulation time in minutes
arrivals = np.random.poisson(lambda_arrival, num_minutes)
services = np.random.exponential(1/lambda_service, num_minutes)

# Simulating the queue process
queue = 0
wait_times = []

for minute in range(num_minutes):
    if queue > 0:
        queue -= services[minute]  # number of customers served
    queue += arrivals[minute]      # new arrivals
    queue = max(queue, 0)          # queue can't be negative
    wait_times.append(queue)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(wait_times)
plt.title('Simulated Queue Length Over Time')
plt.xlabel('Time (minutes)')
plt.ylabel('Queue Length')
plt.show()
```
